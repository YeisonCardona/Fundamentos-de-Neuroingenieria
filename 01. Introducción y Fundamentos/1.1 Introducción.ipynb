## 1.1.3 Procesamiento de Señales

El procesamiento de señales en el contexto de la Neuroingeniería es un área esencial y compleja que implica la análisis, interpretación, y manipulación de señales neurales y biológicas. Su importancia en la Neuroingeniería se refleja en varias aplicaciones clave:

# Actividad

**Objetivo:** Generar una señal sinusoide que simule una onda cerebral, aplicar ruido, y analizar tanto en el dominio del tiempo como en el dominio de la frecuencia. La actividad proporcionará una comprensión básica de cómo se pueden crear, manipular, y examinar señales neurales utilizando Python, enfocándose en la visualización y el análisis espectral mediante la transformación de Fourier.

## Paso 1: Importar las Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Paso 2: Generar una Señal Sinusoide (Simulación de Onda Cerebral)

```python
fs = 1000  # Frecuencia de muestreo en Hz
f = 10     # Frecuencia de la señal en Hz (por ejemplo, onda alfa)
t = np.arange(0, 1, 1/fs) # Vector de tiempo
signal = np.sin(2 * np.pi * f * t) # Señal sinusoide
```

## Paso 3: Añadir Ruido

```python
noise = np.random.normal(0, 0.5, signal.shape) # Ruido gaussiano
signal_noise = signal + noise # Señal con ruido
```

## Paso 4: Transformación de Fourier
Aplicaremos la Transformada Rápida de Fourier (FFT) para observar el contenido de frecuencia de las señales original y con ruido.

```python
from scipy.fft import rfft, rfftfreq

# Transformada de Fourier
fft_original = rfft(signal)
fft_noise = rfft(signal_noise)

# Frecuencias correspondientes
freqs = rfftfreq(len(t), 1/fs)
```

## Paso 5: Visualizar las Señales en el Dominio del Tiempo

```python
plt.figure(figsize=(14, 7))

plt.plot(t, signal, label='Original', zorder=10)
plt.plot(t, signal_noise, label='Con Ruido')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
```

## Paso 6: Visualizar las Señales en el Dominio de la Frecuencia

```python
plt.figure(figsize=(12, 6))

plt.subplot(211)
plt.plot(freqs, np.abs(fft_original), label='Original')
plt.ylabel('Amplitud')
plt.legend()

plt.subplot(212)
plt.plot(freqs, np.abs(fft_noise), label='Con Ruido')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```
